## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty

# SAMPLE OUTPUT
# unit_price	qty	line_subtotal
# 0	1096.80	105	115164.00
# 1	24.95	21	523.95

# Create a line_subtotal column that is the total price of the quantity ordered
df["line_subtotal"] = df["unit_price"] * df["qty"]

# Display output that matches the sample output
df[["unit_price", "qty", "line_subtotal"]].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

# SAMPLE OUTPUT
# 	unit_price	unit_weight	qty	total_weight	shipping_price
# 0	1096.80	7.50	105	787.50	5512.50
# 1	24.95	1.49	21	31.29	312.90
# 2	13.52	1.68	39	65.52	458.64

# Calculate total weight and create column in the dataframe as seen in the sample output 
df["total_weight"] = df["unit_weight"] * df["qty"]

# Calculate shipping price based on requirements and create column as seen in sample output
df["shipping_price"] = df["total_weight"] * df["total_weight"].apply(lambda x: 7 if x > 50 else 10)

# Display first three entries as shown in sample output
df[["unit_price", "unit_weight", "qty", "total_weight", "shipping_price"]].head(3)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

# SAMPLE OUTPUT
# line_subtotal	shipping_price	line_price
# 0	115164.00	5512.50	131839.08
# 1	523.95	312.90	914.26
# 2	527.28	458.64	1077.12

# Add the subtotal and shipping price together, then multiply it by 1 plus the sales tax of 9.25% to get full price. Round to two decimal places.
# Create a new dataframe column called "line_price" and store the value in that column
df["line_price"] = ((df["line_subtotal"] + df["shipping_price"]) * 1.0925).round(2)

# Display output that matches the sample output
df[["line_subtotal", "shipping_price", "line_price"]].head(3)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

# SAMPLE OUTPUT
# first	last	job	phone	email	client_id	order_id	order_date	order_week	order_year	...	unit_price	unit_cost	unit_weight	qty	line_number	line_subtotal	total_weight	shipping_price	line_price	line_cost
# 0	Donald	Harding	Immunologist	793-904-7725x39308	harding.donald.7185@sullivan.com	58515	8953482	2023-04-28	17	2023	...	1096.80	762.71	7.50	105	1	115164.00	787.50	5512.50	131839.08	85597.05
# 1	Tiffany	Myers	Music therapist	201.442.4543x942	myers.t.6537@ferguson-johnson.net	37609	8069089	2023-05-19	20	2023	...	24.95	15.09	1.49	21	0	523.95	31.29	312.90	914.26	629.79
# 2	Shannon	Watson	Immunologist	687.737.9424x8503	swatson8146@payne.net	57113	1902144	2023-01-29	4	2023	...	13.52	7.86	1.68	39	6	527.28	65.52	458.64	1077.12	765.18

# Multiple the unit cost times the quantity of the item, then add the shipping price to get the line cost
# Create a new dataframe column called "line_cost" to store that computed value
df["line_cost"] = (df["unit_cost"] * df["qty"]) + df["shipping_price"]

# Display first three rows in dataframe to match sample output
df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price

# SAMPLE OUTPUT
# 	first	last	job	phone	email	client_id	order_id	order_date	order_week	order_year	...	unit_cost	unit_weight	qty	line_number	line_subtotal	total_weight	shipping_price	line_price	line_cost	line_profit
# 0	Donald	Harding	Immunologist	793-904-7725x39308	harding.donald.7185@sullivan.com	58515	8953482	2023-04-28	17	2023	...	762.71	7.50	105	1	115164.00	787.50	5512.50	131839.08	85597.05	46242.03
# 1	Tiffany	Myers	Music therapist	201.442.4543x942	myers.t.6537@ferguson-johnson.net	37609	8069089	2023-05-19	20	2023	...	15.09	1.49	21	0	523.95	31.29	312.90	914.26	629.79	284.47
# 2	Shannon	Watson	Immunologist	687.737.9424x8503	swatson8146@payne.net	57113	1902144	2023-01-29	4	2023	...	7.86	1.68	39	6	527.28	65.52	458.64	1077.12	765.18	311.94

# Subtract the line cost from the line price to get the line profit
# Create a new column in the dataframe called "line_profit" and store the values in that column
df["line_profit"] = df["line_price"] - df["line_cost"]

# Display the first three rows of the dataframe to match the sample output
df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

# SAMPLE OUTPUT
# 33615: $8377308.52
# 66037: $10259514.79
# 46820: $9743794.36
# 38378: $12906550.87
# 24741: $82268892.02

# Get all orders made by the top 5 clients stored in the list from Part 1
top_5_client_orders = df.loc[df["client_id"].isin(top_client_ids)]

# Sum up the line_profit of ever ordered item by each of the top 5 clients
top_5_client_spend_series = top_5_client_orders.groupby("client_id")["line_price"].sum().round(2)

# Print output to match sample output
for client_id in top_client_ids:
    print(f'{client_id}: ${top_5_client_spend_series[client_id]}')
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

# SAMPLE OUTPUT
# client_id	qty	shipping_price	line_price	line_cost	line_profit
# 0	33615	64313	1828984.89	8377308.52	6175313.91	2201994.61
# 1	66037	43018	1395151.85	10259514.79	7004482.98	3255031.81
# 2	46820	75768	1601448.84	9743794.36	7007191.64	2736602.72
# 3	38378	73667	3429455.40	12906550.87	9634720.98	3271829.89
# 4	24741	239862	5126448.37	82268892.02	45688899.71	36579992.31

# Initialize the summary DataFrame - The first line is unnecessary but the Pylance linter complains if I leave it out
top_5_client_summaries_df = pd.DataFrame()
top_5_client_summaries_df["client_id"] = top_client_ids 

# Get names of columns to create
column_headers = ["qty", "shipping_price", "line_price", "line_cost", "line_profit"]

# Create each column and populate it with the sum of that column in the top 5 client DataFrame for that client_id
for col in column_headers:
    for id in top_client_ids:
        top_5_client_summaries_df.loc[top_5_client_summaries_df["client_id"] == id, col] = \
            top_5_client_orders.loc[top_5_client_orders["client_id"] == id][col].sum().round(2)

# Convert the qty column to an int
top_5_client_summaries_df["qty"] = top_5_client_summaries_df["qty"].astype(int)

# Print output to match sample output
display(top_5_client_summaries_df)
```

```python
# Format the data and rename the columns to names suitable for presentation.

# SAMPLE OUTPUT
# Client ID	Units	Shipping (millions)	Total Revenue (millions)	Total Cost (millions)	Total Profit (millions)
# 0	33615	64313	1.828985	8.377309	6.175314	2.201995
# 1	66037	43018	1.395152	10.259515	7.004483	3.255032
# 2	46820	75768	1.601449	9.743794	7.007192	2.736603
# 3	38378	73667	3.429455	12.906551	9.634721	3.271830
# 4	24741	239862	5.126448	82.268892	45.688900	36.579992

# Define the money columns. 
money_columns = ["shipping_price", "line_price", "line_cost", "line_profit"]

# Define a function that converts a dollar amount to millions.
def currency_format_millions(amount):
    return amount/1000000

# Apply the currency_format_millions function to only the money columns. 
for money_col in money_columns:
    top_5_client_summaries_df[money_col] = top_5_client_summaries_df[money_col].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
new_column_names = {
    "client_id": "Client ID",
    "qty": "Units", 
    "shipping_price": "Shipping (millions)", 
    "line_price": "Total Revenue (millions)", 
    "line_cost": "Total Cost (millions)", 
    "line_profit": "Total Profit (millions)"
}

top_5_client_summaries_df = top_5_client_summaries_df.rename(columns=new_column_names)
top_5_client_summaries_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.

# SAMPLE OUTPUT
# 	Client ID	Units	Shipping (millions)	Total Revenue (millions)	Total Cost (millions)	Total Profit (millions)
# 4	24741	239862	5.126448	82.268892	45.688900	36.579992
# 3	38378	73667	3.429455	12.906551	9.634721	3.271830
# 1	66037	43018	1.395152	10.259515	7.004483	3.255032
# 2	46820	75768	1.601449	9.743794	7.007192	2.736603
# 0	33615	64313	1.828985	8.377309	6.175314	2.201995

top_5_client_summaries_sorted_df = top_5_client_summaries_df.sort_values("Total Profit (millions)", ascending=False)

display(top_5_client_summaries_sorted_df)
```
